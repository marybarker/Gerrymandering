```python
%%time
from helper.setup import *
from helper.pycuda_setup import *

DIV1 = "*" * 80
DIV2 = "#" * 80


gridDims = (1,1,1)
blockDims = (1,1,1)
max_steps = 2
num_sims = np.product([gridDims,blockDims])
rand_seeds = np.arange(num_sims)

num_clrs = 3
num_vtds = 10
vtd_pop = np.random.randint(20,size=num_vtds)

num_vtds = len(vtd_pop)

#If given edge list
edges = [(0,4)
        ,(1,4)
        ,(2,4)
        ,(4,3)
        ,(3,4)        
        ]

#If given neighbor list
"""
nbrs = [[1,4]
       ,[0,4]
       ,[4]
       ,[4]
       ,[0,1,2,3]
       ]
edges = [(v,w) for (v,N) in enumerate nbrs for w in N if v<w]
"""

num_edges = int((num_vtds*(num_vtds-1)/2)*.75)
edges = [np.random.choice(num_vtds, size=2, replace=False) for _ in range(num_edges)]

for (idx,e) in enumerate(edges):
    if(len(e) != 2):
        raise Exception('edge %d: %s is not length 2'%(idx, e))
    if((e[0]%1 != 0) | (e[1]%1 != 0)):
        raise Exception('edge %d: %s contains a non-integer'%(idx, e))


edges = [tuple(sorted(e)) for e in edges if e[0] != e[1]] #make edge (smaller,larger) & remove self-loops
edges = set(edges) #removes duplicates
edges = sorted(list(edges)) #sorts
edges = np.array(edges).astype('uint16')

if num_vtds != edges.max()+1:
    raise Exception('length of population vector != max vtd in edge list.  Is there an isolated vtd?')

vtd0 = edges[:,0]
vtd1 = edges[:,1]
num_edges = len(vtd0)

nbrs0 = [[] for v in range(num_vtds)]
for e in edges:
    nbrs0[e[0]].append(e[1])
    nbrs0[e[1]].append(e[0])
nbrs = [np.array(n).astype('uint16') for n in nbrs0]
del nbrs0

degree = np.array([len(n) for n in nbrs])
nbr_rng = np.insert(np.cumsum(degree),0,0)

print("edges\n%s"%edges)
print("num_edges = %s"%num_edges)
print("vtd0 = %s"%vtd0)
print("vtd1 = %s"%vtd1)
print("num_vtds = %s"%num_vtds)
print("neighbors = %s"%nbrs)
print("degree = %s"%degree)
print("neighbor index range = %s"%nbr_rng)
print(DIV2)
print(DIV2)


#def clr_stats(printout=True):
#    global clr_count, clr_pop, num_uclr#, clr_vtd

def print_stats():
    print("vtd colors")
    print(clr)
    print("#vtds per color")
    print(clr_count)
    print("population per color")
    print(clr_pop)
    print("number of uncolored vtds = %d"%num_uclr)

def recolor(w, c):
    global clr_count, clr_pop, num_uclr
    clr_count[clr[w]] -= 1
    clr_pop[clr[w]] -= vtd_pop[w]
    clr[w] = c
    clr_count[clr[w]] += 1
    clr_pop[clr[w]] += vtd_pop[w]
    num_uclr -= 1

num_sims = 1
clr_arr = num_clrs * np.ones([num_sims,num_vtds]).astype('uint16')

for s in range(num_sims):
    print("Map %d"%m)
    print("Initial coloring - place colors on vtds with smallest degree")
    small_deg_idx = np.argpartition(degree,num_clrs)[:num_clrs]
    clr = num_clrs * np.ones(num_vtds).astype('uint16')
    init_clr = np.random.permutation(num_clrs)
    clr[small_deg_idx] = init_clr

    clr_count = np.bincount(clr).astype('uint16')
    clr_pop = np.bincount(clr,weights=vtd_pop).astype('uint32')
    num_uclr = clr_count[-1]
    print_stats()

    Steps = num_uclr
    for step in range(Steps):
        clr_pop_srt = np.argsort(clr_pop[:-1])
        print(DIV2)
        print(DIV2)
        print("STEP %d"%step)        
        print_stats()
        print(clr_pop_srt)
        print(DIV1)
        for c in clr_pop_srt:
            c_idx = np.where(clr==c)[0]
            c_idx.shuffle()
            print("Trying to add vtd to color %d.  Here are vtds with that color (in random order):"%c)
            print(c_idx)
            found = False
            for v in c_idx:
                M = nbrs[v]
                N = M[clr[M] == num_clrs]
                print("Trying to color a neighbor of %d.  Here are the uncolored neighbors."%v)
                print(N)
                if(len(N) == 0):
                    print("All neighbors of %d are already colored.  Trying next vtd with color %d."%(v, c))
                else:
                    w = np.random.choice(N,1)
                    print("vtd %d is NOT already colored.  Coloring it %d."%(w,c))
                    recolor(w, c)
                    found = True
                    break
            if(found == False):
                print("Unable to add ANY neighboring vtds to color %d because all neighbors are already colored.  Trying next color."%c)
                print(DIV1)
                print(DIV1)
            if(found == True):            
                break
        if(found == False):
            c = clr_pop_srt[0]
            nbr = np.where(clr==num_clrs)[0]
            w = nbr[np.argmin(degree[nbr])]
            print("Can't add ANY neighboring vtds to ANY color.  Putting color %d on a vtd %d - it has smallest degree of the uncolored vtds."%(c,w))
            recolor(w, c)
        print(DIV2)
        print_stats()
    print(DIV2)
    print(DIV1)
    print(DIV2)
    print(DIV1)
    clr_arr[s,:] = clr
```
